```python
import pandas
import geopandas
import h3
from shapely import wkt
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
import h3
```

```python
rj_instances = pandas.read_csv("../data/cvrp-instances-1.0/train/cvrp-0-rj-0.csv")
rj_area = pandas.read_csv("/media/gegen07/Expansion/data/rj_interpolated_res_6.csv")
```

```python
rj_area['geometry'] = rj_area.geometry.apply(wkt.loads)
rj_area = geopandas.GeoDataFrame(rj_area, geometry="geometry")
```

```python
rj_area["h3_boundary"] = rj_area.centroid.apply(lambda x: h3.geo_to_h3(x.y, x.x, 6))
```

```python
bag = pandas.DataFrame(rj_instances.groupby("h3_boundary").count()["id"]).reset_index()
```

```python
bag.columns = ["h3_boundary", "count"]
```
